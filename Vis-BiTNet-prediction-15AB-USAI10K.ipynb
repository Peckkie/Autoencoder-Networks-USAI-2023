```python
import os
import numpy as np
import pandas as pd

# load BiTNet model
import pickle
os.chdir('/media/tohn/SSD/ModelTrainByImages/R2_1/models/')
filename = "modelrf_EffNet5_b5_1FC_cross_15AB_15AB.pkl"
with open(filename, 'rb') as file:
    model = pickle.load(file)

# open csv
dataframe = pd.read_csv('/media/tohn/HDD/Dataset_USAI10K/csv/FV_USAI10K_Doctor-all_2013-2023_5FP_15AB_all.csv')
dataframe = dataframe.drop(['Unnamed: 0'], axis=1)

X_test = dataframe.iloc[:,0:2048]
y_test = dataframe.iloc[:,2049]
print(X_test.shape)
print(y_test.shape)
```

```python
labels = {0: 'AB01', 1: 'AB02', 2: 'AB03', 3: 'AB04', 4: 'AB05', 5: 'AB06', 6: 'AB07', 7: 'AB081', 8: 'AB082', 9: 'AB083', 
          10: 'AB09', 11: 'AB10', 12: 'AB11', 13: 'AB12', 14: 'Normal'}
```

```python
# Old CSV file that before fv-extracted from the EffNet model
df = pd.read_csv('/home/kannika/code/USAI_Doctor-all_2013-2023_dummy_pathcrop_AB_ver2_5FPViT.csv') 
df = df.drop(['Unnamed: 0', 'Unnamed: 0.1', 'Unnamed: 0.1.1'], axis=1)
df.head()
```

```python
# The CSV file that FV result from EffNet model
dataframe.head()
```

```python
img_path=df['img_path_new'].tolist()
```

```python
res_list = list()
# y_pred = model.predict(X_test)
y_prob = model.predict_proba(X_test)
for i in range(0,len(y_prob)):
    result1 = y_prob[i]
    res_list.append({'img_path':img_path[i] , 'AB01':result1[0], 'AB02': result1[1], 'AB03': result1[2], 'AB04': result1[3], 
                     'AB05': result1[4], 'AB06': result1[5], 'AB07': result1[6], 'AB081': result1[7], 'AB082': result1[8], 
                     'AB083': result1[9], 'AB09': result1[10],'AB10': result1[11], 'AB11': result1[12], 'AB12': result1[13], 
                     'Normal': result1[14]})
```

```python
resdf = pd.DataFrame(res_list)
resdf.head()
```

```python
resdf['GroupAB_Abnormolity'] = dataframe['GroupAB_Abnormolity']
# resdf['ab_class'] = dataframe['ab_class']
resdf['category_5FPvit'] = dataframe['category_5FPvit']
resdf['Prob_5FPvit'] = dataframe['Prob_5FPvit']
resdf['category_BiTNet'] = dataframe['category_BiTNet']
resdf['Prob_BiTNet'] = dataframe['Prob_BiTNet']
resdf.head()
```

## + Finding relative order of elements in prob. of 15AB classes.

```python
res_list = list()
img_path=df['img_path_new'].tolist()
y_prob = model.predict_proba(X_test)
for i in range(0,len(y_prob)):
    test_list = y_prob[i]
    temp = sorted(test_list, reverse=True)   
    result1 = [temp.index(i) for i in test_list]
    res_list.append({'img_path':img_path[i] , 'AB01':result1[0], 'AB02': result1[1], 'AB03': result1[2], 'AB04': result1[3], 
                     'AB05': result1[4], 'AB06': result1[5], 'AB07': result1[6], 'AB081': result1[7], 'AB082': result1[8], 
                     'AB083': result1[9], 'AB09': result1[10],'AB10': result1[11], 'AB11': result1[12], 'AB12': result1[13], 
                     'Normal': result1[14]})
resdf_order = pd.DataFrame(res_list)
resdf_order.head() 
```

## + All Top

```python
pred_list = list()
prob_list = list()

y_prob = model.predict_proba(X_test)

for i in range(0,len(y_prob)):
    result = y_prob[i]
    
    re_class = list()
    re_prop = list()
    
    while len(re_class) <2:
        maxx = np.argmax(result)
        re_class.append(labels[maxx])
        re_prop.append(result[maxx])
        result = np.where(result==result[maxx], 0, result)
    pred_list.append(re_class)
    prob_list.append(re_prop)
    
resdf['Top2_category_BiTNet'] = pred_list
resdf.head()
```
