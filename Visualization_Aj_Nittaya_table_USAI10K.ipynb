```python
# open csv
import pandas as pd
dataframe = pd.read_csv('/media/tohn/HDD/Dataset_USAI10K/csv/USAI10K188209_BiTNet_Predicted_All.csv')
dataframe = dataframe.drop(['Unnamed: 0'], axis=1)
# print(dataframe.shape)

# PID and Doctor ID
PID_df = pd.read_csv('/media/tohn/HDD/Dataset_USAI10K/csv/USAI10K_Doctor-all_2013-2023_5FP_15AB_ALL.csv')
PID_df = PID_df.drop(['Unnamed: 0'], axis=1)


dataframe['Doctor_ID'] =  PID_df['Doctor']
dataframe['PID'] =  PID_df['pid']
dataframe  = dataframe.astype({'Doctor_ID': 'object', 'PID': 'object'})
```

```python
dataframe
```

```python
did = list(set(dataframe['Doctor_ID']))
did
```

```python
# only select Nittaya doctor
dataframe_12 = dataframe[dataframe['Doctor_ID'] == 12]
dataframe_12
```

```python
# finding  Prob_BiTNet_of15AB
dataframe_12['Prob_BiTNet_ofAbNor'] = dataframe_12['Prob_BiTNet']
prob_max = []

for i in range(len(dataframe_12)):
    prob = dataframe_12.iloc[i,1:16].tolist()
    prob_max.append(max(prob))
    
dataframe_12['Prob_BiTNet_of15AB'] = prob_max
```

```python
# finding  Result_BiTNet_Currect

GroupAB_Abnormolity = dataframe_12['GroupAB_Abnormolity'].tolist()
category_BiTNet = dataframe_12['category_BiTNet'].tolist()
res =[]

for i in range(len(GroupAB_Abnormolity)):
    
    cla = GroupAB_Abnormolity[i].split('[')[1].split(']')[0].split(', ')
    if f"'{category_BiTNet[i]}'" in cla:
        res.append(1)

    else:
        res.append(0)
        
dataframe_12['Result_BiTNet_Currect'] = res
```

```python
dataframe_12.groupby(['Result_BiTNet_Currect'])['img_path'].count().reset_index()
```

```python
dataframe_12 = dataframe_12[['img_path', 'AB01', 'AB02', 'AB03', 'AB04', 'AB05', 'AB06', 'AB07',
       'AB081', 'AB082', 'AB083', 'AB09', 'AB10', 'AB11', 'AB12', 'Normal',
       'GroupAB_Abnormolity', 'category_5FPvit', 'Prob_5FPvit',
       'category_BiTNet', 'Result_BiTNet_Currect', 'Prob_BiTNet_ofAbNor', 'Prob_BiTNet_of15AB', 'Doctor_ID', 'PID']]
```

# ðŸ”—sort by the AI-confidence

```python
dataframe_12_ofAbNor = dataframe_12.sort_values(by='Prob_BiTNet_ofAbNor', ascending=False)
dataframe_12_ofAbNor
```

```python
dataframe_12_ofAbNor = dataframe_12_ofAbNor.astype('object')
```

```python
dataframe_12_of15AB = dataframe_12.sort_values(by='Prob_BiTNet_of15AB', ascending=False)
dataframe_12_of15AB
```

```python
dataframe_12_of15AB = dataframe_12_of15AB.astype('object')
```
